# <a name="04intro">Introduction to Probability</aa>

------------------------------------------------------------------------

Statistical inference is the process of using data from a sample to
describe characteristics of the population. Our conclusions are going to
be based on randomness in the sampling process, and we will need to
account for some uncertainty in our predictions. Statistics relies on
theory and applications from probability to help quantify the
uncertainty in our models and estimates.

-   <span style="color: dodgerblue;">**If we know the make up of a
    population, we can use probability to predict the likelihood that
    certain outcomes occur.**</span>
    -   If we know a six-sided is fair, what is the probability of
        rolling the die 10 times and getting exactly 4 rolls that land
        on 1?
-   <span style="color: dodgerblue;">**We use statistics to go in the
    reverse direction. Namely, how can we make predictions about a
    population from a random sample of data.**</span>
    -   We want to determine whether or not a six-sided die is fair. We
        roll the die 100 times and get 40 rolls that land on 1. Is the
        die fair?
    -   We will need to use probability to answer statistical questions
        and make inferences about a population.

## <a name="04sample-space">Sample Space, Outcomes and Events</aa>

------------------------------------------------------------------------

A <span style="color: dodgerblue;">**statistical experiment or
observation**</span> is any random activity that results in a definite
outcome.

-   The <span style="color: dodgerblue;">**sample space**</span>
    $\color{dodgerblue}{\Omega}$ is the set of all possible outcomes of
    an experiment.
-   An <span style="color: dodgerblue;">**outcome**</span>
    $\color{dodgerblue}{\omega}$, is a result from an experiment or
    observation.
-   An <span style="color: dodgerblue;">**event**</span>,
    $\color{dodgerblue}{A}$, is a collection of one or more outcomes
    from an experiment or observation.

## <a name="04equal-likelihood">Probabilities with Equally Likely Outcomes</aa>

------------------------------------------------------------------------

For finite sample spaces, we often use counting to determine
probabilities. A special case which we will encounter often is when each
outcome in the sample space $\Omega$ is equally likely to occur, and
therefore

$$ P(A) = \frac{\mbox{Number of outcomes in event $A$}}{\mbox{Total number of outcomes in $\Omega$}} = \frac{|A|}{|\Omega|}.$$

We use the notation $P(A)$ to denote the probability that event $A$
occurs.

-   Probabilities are proportions between $0$ (impossible to occur) and
    $1$ (certain to occur) that we typically represent as decimals or
    fractions.
-   The probability $P: \Omega \rightarrow \mathbb{R}$ is a function
    that maps each event in $\Omega$ to a number in the interval
    $0 \leq P(A) \leq 1$.
-   Sometimes we convert the proportion to a percentage when giving a
    probability.

# <a name="O4notation">Simple and Compound Probabilities</aa>

------------------------------------------------------------------------

Let $A$ and $B$ denote two events in sample space $\Omega$, then

-   $\color{dodgerblue}{P(A)}$ is the probability that event $A$ occurs.
-   $\color{dodgerblue}{P(A^C) = P(\bar{A}) = P(A')}$ is the probability
    that event $A$ <span style="color: dodgerblue;">**does NOT
    occur**</span>.
-   The notation $\color{dodgerblue}{A^C}$,
    $\color{dodgerblue}{\bar{A}}$, or $\color{dodgerblue}{A'}$ are used
    to denote the <span style="color: dodgerblue;">**complement**</span>
    of $A$.
-   $\color{dodgerblue}{P(A \cap B)}$ is the probability that events $A$
    <span style="color: dodgerblue;">**and**</span> $B$ both occur.
-   $\color{dodgerblue}{P(A \cup B)}$ is the probability that either
    event $A$ <span style="color: dodgerblue;">**or**</span> event $B$
    occurs.
-   $\color{dodgerblue}{P(B \ | \ A )}$ is the
    <span style="color: dodgerblue;">**conditional probability**</span>
    that event $B$ occurs <span style="color: dodgerblue;">**given
    that**</span> event $A$ occurs.
-   $\color{dodgerblue}{P(A - B)}$ is the probability that
    <span style="color: dodgerblue;">**event A occurs and event B does
    not occur**</span>.

# <a name="04prop">Properties of Probability</aa>

------------------------------------------------------------------------

We can generalize the calculations from the previous study on people
arrested for small quantities of marijuana to obtain the following
results:

Let $A$ and $B$ denote two events in sample space $\Omega$, then

-   <span style="color: dodgerblue;">**Additive property**</span>:
    $P(A \cup B) = P(A) + P(B) - P(A \cap B)$.
-   <span style="color: dodgerblue;">**Bayes’ Theorem**</span>:
    $\displaystyle P(B | A) = \frac{P(A \cap B)}{P(A)}$
-   <span style="color: dodgerblue;">**Multiplicative property**</span>:
    $P(A \cap B) = P(A) \cdot P(B | A)$
-   <span style="color: dodgerblue;">**Complement property**</span>:
    $P(A^C) = 1 - P(A)$

## <a name="04def-ind">Definition of Independent Events</aa>

------------------------------------------------------------------------

Two events $A$ and $B$ are
<span style="color: dodgerblue;">**independent**</span> if the
occurrence of one has no effect on the occurrence of the other:

$$ P(B) = P(B \ | \ A) \quad \mbox{or} \quad P(A) = P(A \ | \ B).$$

**Special case:** If events $A$ and $B$ are independent events then we
have $P(A \cap B) = P(A)P(B)$.

# <a name="04excl">Mutually Exclusive Events</aa>

------------------------------------------------------------------------

Two events $A$ and $B$ are <span style="color: dodgerblue;">**mutually
exclusive**</span> (or
<span style="color: dodgerblue;">**disjoint**</span>) if they cannot
occur at the same time, and therefore $P(A \cap B) = 0$.

**Special case:** If events $A$ and $B$ are disjoint then we have
$P(A \cup B) = P(A)+P(B)$.

# <a name="04prob-dist">Probability Distributions</aa>

------------------------------------------------------------------------

A function $P$ that assigns a real number $P(A)$ to each event $A$ is a
<span style="color: dodgerblue;">**probability distribution**</span> or
a <span style="color: dodgerblue;">**probability measure**</span> if it
satisfies the following three axioms:

-   $P(A) \geq 0$ for all $A$.
-   $P(\mbox{full sample space})=P(\Omega) = 1$.
-   If $A$ and $B$ are disjoint, then

$$P \left( A \cup B \right) = P(A) + P(B).$$

------------------------------------------------------------------------

$\Omega$ is a collection of disjoint sets such that

## <a name="04state-bayes">Statement of Bayes’ Theorem</aa>

------------------------------------------------------------------------

Let $A_1$, $A_2$, $\ldots , A_k$ be a partition of $\Omega$ such that
$P(A_i)>0$ for each $i$. If $B$ is any event with $P(B)>0$, then for
each $i=1, \ldots k$, we have Then for any event $B$,

$$P(A_i \mid B) = \frac{P(A_i \cap B) }{P(B)} = \frac{P(B \mid A_i) P(A_i)}{\sum_{j=1}^k P(B \mid A_j)P(A_j)}.$$
